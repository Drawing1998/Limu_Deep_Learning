```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l
```

# 1. 生成数据集

```python
def synthetic_data(w, b, num_examples):
    """生成y=Xw+b+噪声"""
    X = torch.normal(0, 1, (num_examples, len(w)))   # 生成均值为0，方差为1，num_examples个样本，每个样本包含len(w)个特征的数据集X
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)    # 给y加上一个噪声矩阵，噪声矩阵的形状与y相同，每个元素都服从均值为0，方差为0.01的高斯分布
    return X, y.reshape((-1, 1))     # reshape成1列


true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)

d2l.set_figsize()
d2l.plt.scatter(features[:, 1].detach().numpy(), labels.detach().numpy(), 1)
```

# 2. 读取数据集

每次读取batch_size小批量的数据。

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))    # 下标
    random.shuffle(indices)    # 这些样本是随机读取的，没有特定的顺序
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]   # 带yield的函数是一个生成器，会迭代的返回生成的数据


batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)    # 打印一个小批量的数据
    break
```

# 3. 初始化模型参数

```python
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)    # requires_grad=True 需要计算梯度
b = torch.zeros(1, requires_grad=True)     # 将偏置项初始化为0
```

# 4. 定义模型

广播机制：当用一个向量加一个标量时，标量会被加到向量的每个分量上。

```python
def linreg(X, w, b):
    """线性回归模型"""
    return torch.matmul(X, w) + b
```

# 5. 定义损失函数

```python
def squared_loss(y_hat, y):
    """均方损失"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2 / len(y)
```

# 6. 定义优化算法——小批量随机梯度下降

```python
def sgd(params, lr, batch_size):
    """小批量随机梯度下降"""
    with torch.no_grad():       # 在进行参数更新时不要计算梯度
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()    # 更新完参数之后把梯度清零，因为每次更新完参数之后pytorch不会自动将梯置为为零
```

# 7. 训练

在每次迭代中，我们读取一小批量训练样本，并通过我们的模型来获得一组预测。 计算完损失后，我们开始反向传播，存储每个参数在损失函数中的梯度。 最后，我们调用优化算法sgd来更新模型参数。

```python
lr = 0.03         # 学习率
num_epochs = 3    # 迭代3次
net = linreg      # 模型
loss = squared_loss

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)  # X和y的小批量损失
        # 因为l形状是(batch_size,1)，而不是一个标量。l中的所有元素被加到一起，
        # 并以此计算关于[w,b]的梯度
        l.sum().backward()    # 计算损失函数l关于[w,b]的梯度
        sgd([w, b], lr, batch_size)  # 使用参数的梯度更新参数
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}')
```
